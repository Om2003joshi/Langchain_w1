### 1.Text loader


```python
from langchain_community.document_loaders import TextLoader
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from dotenv  import load_dotenv
import os
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader, WebBaseLoader
from bs4 import BeautifulSoup
```

```python
load_dotenv()

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)
```

```python
p1 =PromptTemplate(
    template="write a summary for the following poem - \n {poem}",
    input_variables=['poem']
)
```

```python
parser=StrOutputParser()
```

```python
loader=TextLoader('cricket.txt',encoding='utf-8')
```

```python
doc=loader.load()
```

```python
print(doc)
```

```python
print(type(doc))
```

```python
print(len(doc))
```

```python
print(doc[0].page_content)
```

```python
print(doc[0].metadata)
```

```python
chian= p1 | llm | parser
```

```python
chian.invoke({"poem":doc[0].page_content})
```

### 2.PyPDFLoader

```python
import os
print(os.listdir("books"))
```

```python
loader = DirectoryLoader(
    path='books',
    glob='*.pdf',
    loader_cls=PyPDFLoader
)
```
