```python
class nllm:
    def __init__(self):
        print('llm created')
        
    def predict(self,prompt):
        responce_list=[
        'delhi is capital of indaia',
        'ipl is cricket leauge',
        'ai stands gor artificial intelligence'    
        ]
        return{'responce':random.choice(responce_list)}
        
    
```

```python
llm=nllm()
```

```python
class nprompttemplate:
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables
        
    def format(self,input_dict):
        return self.template.format(**input_dict)  
```

```python
template=nprompttemplate(
    template='write a{lenght} poem about {topic}',
    input_variables=['lenght','topoic']
)
```

```python
prompt=template.format({'lenght':'short','topic':'imdia'})
```

```python
llm.predict(prompt)
```
