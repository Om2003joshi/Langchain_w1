```python
import os
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
# tool creation...
```

```python
# tool creater
@tool
def mutiply (a: int,b: int) ->int:
    """given 2 numbers a and this tool returns theur products"""
    return a*b
```

```python
print(mutiply.invoke({'a':3,'b':4}))
```

```python
mutiply.name
```

```python
mutiply.description
```

```python
mutiply.args
```

```python
# Load environment variables from .env
load_dotenv()

# Initialize the LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)
```

```python
llm_with_tool=llm.bind_tools([mutiply])
```

### Tool calling

```python
llm_with_tool.invoke("hi how are you?")
```

```python
llm_with_tool.invoke("can you mulitply 3 with 10").tool_calls[0]
```

```python
re=llm_with_tool.invoke("can you mulitply 3 with 10")
```
