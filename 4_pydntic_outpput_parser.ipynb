```python
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel,Field
from langchain_core.prompts import PromptTemplate
```

```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="moonshotai/kimi-k2-instruct-0905",  # âœ… spelling fixed
    api_key=os.getenv("GROQ_API_KEY"),
    temperature=0.3,
)

response = llm.invoke("Hi")
print(response.content)
```

```python
class person(BaseModel):
    name:str=Field(description="name of the the person")
    age:int=Field(gt=18,description="age of the person")
    city:str=Field(description="name of the city the person belongs to")
```

```python
parser=PydanticOutputParser(pydantic_object=person)
```

```python
t1=PromptTemplate(
    template="generate the name ,age and city of a fictional {place} person \n {fromat_instruction}",
    input_variables=['place'],
    partial_variables={'fromat_instruction':parser.get_format_instructions()}
)
```

```python
p1=t1.invoke({'place':'indian'})
```

```python
re=llm.invoke(p1)
```

```python

print(llm.invoke(p1))
```
