```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

# Free embedding model
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create empty Chroma DB
vectorstore = Chroma(
    collection_name="demo",
    embedding_function=embedding
)
```

```python
from langchain_core.documents import Document

docs = [
    Document(page_content="Virat Kohli is an Indian cricketer."),
    Document(page_content="MS Dhoni was the captain of India."),
    Document(page_content="Cricket is very popular in India."),
]

vectorstore.add_documents(docs)
```

```python
from langchain_community.retrievers import WikipediaRetriever

wiki_retriever = WikipediaRetriever(top_k_results=2)

results = wiki_retriever.invoke("Virat Kohli")
print(results)
```

```python
mmr_retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 2}
)

results = mmr_retriever.invoke("Indian cricket captain")
print(results)
```
