```python
from langchain_google_genai import  ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_core.runnables import Runnable,RunnableParallel,RunnableSequence,RunnablePassthrough,RunnableLambda
import os
```

```python
load_dotenv()

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)

# Send a message
response = llm.invoke("Hi")
print(response.content)
```

```python
def word_count(text):
    return len(text.split())
```

```python
p1=PromptTemplate(
    template="write a joke about {topic}",
    input_variables=['topic']
)
```

```python
parser=StrOutputParser()
```

```python
joke_gen_chain=RunnableSequence(p1, llm , parser)
```

```python
parallel_chain=RunnableParallel({
    "joke": RunnablePassthrough(),
    "word_count":RunnableLambda(word_count)
})
```

```python
final_chain=RunnableSequence(joke_gen_chain, parallel_chain)
```

```python
print(final_chain.invoke({"topic":"ai"}))
```

```python
result=final_chain.invoke({"topic":"ai"})
```

```python
final_reslut="""{} \n word count - {}""".format(result['joke'],result['word_count'])
```
