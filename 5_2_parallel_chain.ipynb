### parallel chain

```python
# user gives txt or documentaion 
# so we generate notes and quize
# from differnte model 
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import Runnable,RunnableParallel
```

```python
from dotenv import load_dotenv
import os
from langchain_google_genai import ChatGoogleGenerativeAI

# Load environment variables from .env
load_dotenv()

# Initialize the LLM
llm1 = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)

# Send a message
response = llm1.invoke("Hi")
print(response.content)

```

```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq

load_dotenv()

llm2 = ChatGroq(
    model="moonshotai/kimi-k2-instruct-0905",  # âœ… spelling fixed
    api_key=os.getenv("GROQ_API_KEY"),
    temperature=0.3,
)

response = llm2.invoke("Hi")
print(response.content)
```

```python
p1=PromptTemplate(
    template="generate short and simple notes from the following text \n {text}",
    input_variables=['text']
    )
```

```python
p2=PromptTemplate(
    template="generate 5 shorts question answers from the following text \n {text}",
    input_variables=['text']
    )
```

```python
p3=PromptTemplate(
    template="merges the provided notes and quiz into a single document \n notes-> {notes} and quiz->{quiz}",
    input_variables=['notes','quiz']
    )
```

```python
parser=StrOutputParser()
```

```python
parallel_chain=RunnableParallel({
    'notes':p1 | llm1 | parser,
    'quiz':p2 | llm2 | parser
})
```

```python
merge_chain=p3 | llm1 | parser
```

```python
chian=parallel_chain | merge_chain
```

```python
text="""Support Vector Machine (SVM)

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression problems. The main idea of SVM is to find an optimal decision boundary (hyperplane) that separates data points of different classes with the maximum margin.

Key Concept

SVM chooses the hyperplane such that the distance between the hyperplane and the nearest data points from each class is maximized. These nearest data points are called support vectors, and they play a crucial role in defining the position of the hyperplane.

Margin

The margin is the distance between the hyperplane and the closest data points of both classes. A larger margin leads to better generalization and lower chances of overfitting.

Kernel Trick

When data is not linearly separable, SVM uses the kernel trick to transform data into a higher-dimensional space where separation becomes possible. Common kernels include:

Linear

Polynomial

Radial Basis Function (RBF)

Sigmoid

Types of SVM

Linear SVM: Used when data is linearly separable.

Non-linear SVM: Uses kernels for complex data.

Support Vector Regression (SVR): Used for regression problems.

Advantages

Effective in high-dimensional spaces

Works well with small and medium-sized datasets

Robust to overfitting due to margin maximization

Disadvantages

Computationally expensive for large datasets

Choice of kernel and parameters is critical

Less interpretable compared to simpler models

Applications

SVM is widely used in:

Text classification

Image recognition

Bioinformatics

Handwriting recognition"""
```

```python
chian.invoke({"text":text})
```
