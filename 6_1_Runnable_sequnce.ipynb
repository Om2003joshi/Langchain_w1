```python
from langchain_google_genai import  ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_core.runnables import Runnable,RunnableParallel,RunnableSequence
import os
```

```python
load_dotenv()
```

```python
p1=PromptTemplate(
    template="write a joke about {topic}",
    input_variables=['topic']
)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)

# Send a message
response = llm.invoke("Hi")
print(response.content)
```

```python
parser=StrOutputParser()
```

```python
p2=PromptTemplate(
    template="explain the following - {text}",
    input_variables=['text']
)
```

```python
chain=RunnableSequence(p1, llm, parser, p2, llm, parser)
```

```python
print(chain.invoke({"topic":"ai"}))
```
