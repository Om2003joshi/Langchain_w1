```python
from langchain_google_genai import  ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_core.runnables import Runnable,RunnableParallel,RunnableSequence,RunnablePassthrough,RunnableLambda,RunnableBranch
import os
```

```python
load_dotenv()

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",                   # correct parameter
    google_api_key=os.environ["GOOGLE_API_KEY"],  # correct parameter
    temperature=0
)

# Send a message
response = llm.invoke("Hi")
print(response.content)
```

```python
p1=PromptTemplate(
    template="write a detailed report on {topic}",
    input_variables=['topic']
)
```

```python
p2=PromptTemplate(
    template="summrize the following text \n {topic}",
    input_variables=['topic']
)
```

```python
parser=StrOutputParser()
```

```python
report_gen_chain=p1 | llm | parser
```

```python
branch_chain= RunnableBranch(
   ( lambda x: len(x.split())>300,p2 | llm | parser),
   RunnablePassthrough()
)
```

```python
final_chain=report_gen_chain | branch_chain
```

```python
final_chain.invoke({"topic":"russia vs ukraine"})
```
